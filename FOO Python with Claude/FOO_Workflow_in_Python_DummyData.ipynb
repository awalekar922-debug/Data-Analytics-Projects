**Reading and Exploring Data**

```python
# Read both CSV files
system_a = pd.read_csv("system_a.csv")
system_b = pd.read_csv("system_b.csv")
```

```python
print("\n" + "=" * 40)
print("SYSTEM B - EXPLORATION")
print("=" * 40)

print(system_b.head())
print(system_b.shape)

print(system_b.columns.tolist())
print(system_b.dtypes)
print(system_b.isnull().sum())
```

**Cleaning the Data**

```python
# ---- CLEANING SYSTEM A ----
print("=" * 40)
print("CLEANING SYSTEM A")
print("=" * 40)

# 1. Make a copy first - always work on a copy, never the original
system_a_clean = system_a.copy()
system_b_clean = system_b.copy()

# 2. Check for duplicates
print("\n📌 Duplicate rows in System A:")
print(system_a_clean.duplicated().sum())

print("\n📌 Duplicate rows in System B:")
print(system_b_clean.duplicated().sum())

# 3. Remove duplicates if any
system_a_clean = system_a_clean.drop_duplicates()
system_b_clean = system_b_clean.drop_duplicates()

# 4. Clean column names - remove spaces, make lowercase
system_a_clean.columns = system_a_clean.columns.str.strip().str.lower()
system_b_clean.columns = system_b_clean.columns.str.strip().str.lower()

print("\n📌 Cleaned Column Names System A:")
print(system_a_clean.columns.tolist())

# 5. Check NAV for any negative or zero values - not valid
print("\n📌 Invalid NAV values in System A (zero or negative):")
print(system_a_clean[system_a_clean["nav"] <= 0])

print("\n📌 Invalid NAV values in System B (zero or negative):")
print(system_b_clean[system_b_clean["nav"] <= 0])

# 6. Flag inactive funds
print("\n📌 Inactive Funds in System A:")
print(system_a_clean[system_a_clean["status"] == "Inactive"])

# 7. Final clean data shape
print("\n📌 System A clean shape:", system_a_clean.shape)
print("📌 System B clean shape:", system_b_clean.shape)

print("\n✅ Cleaning Complete!")
```

**Filter and Isolate**

```python
print("=" * 40)
print("FILTERING & ISOLATING DATA")
print("=" * 40)

#1. Filter by asset class
print("\n📌 Equity Trades Only:")
equity = system_a_clean[system_a_clean["asset_class"] == "Equity"]
print(equity)

print("\n📌 Fixed Income Trades Only:")
fixed_income = system_a_clean[system_a_clean["asset_class"] == "Fixed Income"]
print(fixed_income)

print("\n📌 ETF Trades Only:")
etf = system_a_clean[system_a_clean["asset_class"] == "ETF"]
print(etf)

print("\n📌 REIT Trades Only:")
reit = system_a_clean[system_a_clean["asset_class"] == "REIT"]
print(reit)

# 2. Filter Active vs Inactive
print("\n📌 Active Trades Only:")
active = system_a_clean[system_a_clean["status"] == "Active"]
print(active)

print("\n📌 Inactive Trades Only:")
inactive = system_a_clean[system_a_clean["status"] == "Inactive"]
print(inactive)

# 3. Filter Buy vs Sell
print("\n📌 Buy Trades Only:")
buys = system_a_clean[system_a_clean["trade_type"] == "Buy"]
print(buys)

print("\n📌 Sell Trades Only:")
sells = system_a_clean[system_a_clean["trade_type"] == "Sell"]
print(sells)

# 4. Filter by NAV threshold
print("\n📌 High Value Trades (NAV above 150):")
high_nav = system_a_clean[system_a_clean["nav"] > 150]
print(high_nav)

print("\n📌 Low Value Trades (NAV below 100):")
low_nav = system_a_clean[system_a_clean["nav"] < 100]
print(low_nav)

# 5. Multiple conditions together
print("\n📌 Active Equity Buy Trades:")
active_equity_buy = system_a_clean[
    (system_a_clean["asset_class"] == "Equity") &
    (system_a_clean["status"] == "Active") &
    (system_a_clean["trade_type"] == "Buy")
]
print(active_equity_buy)

# 6. Count summary of each filter
print("\n" + "=" * 40)
print("FILTER SUMMARY")
print("=" * 40)
print(f"Total Trades:          {len(system_a_clean)}")
print(f"Equity Trades:         {len(equity)}")
print(f"Fixed Income Trades:   {len(fixed_income)}")
print(f"ETF Trades:            {len(etf)}")
print(f"REIT Trades:           {len(reit)}")
print(f"Active Trades:         {len(active)}")
print(f"Inactive Trades:       {len(inactive)}")
print(f"Buy Trades:            {len(buys)}")
print(f"Sell Trades:           {len(sells)}")
print(f"High NAV (>150):       {len(high_nav)}")
print(f"Low NAV (<100):        {len(low_nav)}")

print("\n✅ Filtering Complete!")
```

**Reconciliation**

```python
print("=" * 40)
print("RECONCILIATION - SYSTEM A vs SYSTEM B")
print("=" * 40)

# 1. Merge both systems on Trade_ID and Fund
comparison = system_a_clean.merge(
    system_b_clean,
    on=["trade_id", "fund"],
    suffixes=("_sysA", "_sysB")
)

# 2. Compare NAV values
comparison["nav_match"] = comparison["nav_sysA"] == comparison["nav_sysB"]

# 3. Calculate the difference
comparison["nav_difference"] = comparison["nav_sysA"] - comparison["nav_sysB"]

# 4. Show full comparison
print("\n📌 Full Comparison Table:")
print(comparison[["trade_id", "fund", "asset_class_sysA",
                   "nav_sysA", "nav_sysB",
                   "nav_match", "nav_difference"]])

# 5. Isolate mismatches only
mismatches = comparison[comparison["nav_match"] == False]

print("\n📌 Mismatches Found:")
print(mismatches[["trade_id", "fund", "asset_class_sysA",
                   "nav_sysA", "nav_sysB", "nav_difference"]])

# 6. Isolate clean matches
matches = comparison[comparison["nav_match"] == True]
print("\n📌 Clean Matches:")
print(matches[["trade_id", "fund", "asset_class_sysA",
                "nav_sysA", "nav_sysB", "nav_match"]])

# 7. Summary
print("\n" + "=" * 40)
print("RECONCILIATION SUMMARY")
print("=" * 40)
print(f"Total Trades Compared:   {len(comparison)}")
print(f"Matched Trades:          {len(matches)} ✅")
print(f"Mismatched Trades:       {len(mismatches)} ❌")
print(f"Match Rate:              {round(len(matches)/len(comparison)*100, 1)}%")

# 8. Flag severity of mismatch
print("\n📌 Mismatch Severity:")
mismatches_copy = mismatches.copy()
mismatches_copy["severity"] = mismatches_copy["nav_difference"].apply(
    lambda x: "🔴 High" if abs(x) > 5 else "🟡 Low"
)
print(mismatches_copy[["trade_id", "fund", "nav_sysA",
                         "nav_sysB", "nav_difference", "severity"]])

print("\n✅ Reconciliation Complete!")
```

**Summary Report**

```python
print("=" * 40)
print("SUMMARY REPORT")
print("=" * 40)

# 1. Trade count by Asset Class
print("\n📌 Trade Count by Asset Class:")
asset_class_count = system_a_clean.groupby("asset_class")["trade_id"].count()
print(asset_class_count)

# 2. Total NAV by Asset Class
print("\n📌 Total NAV by Asset Class:")
asset_class_nav = system_a_clean.groupby("asset_class")["nav"].sum().round(2)
print(asset_class_nav)

# 3. Average NAV by Asset Class
print("\n📌 Average NAV by Asset Class:")
asset_class_avg = system_a_clean.groupby("asset_class")["nav"].mean().round(2)
print(asset_class_avg)

# 4. Trade count by Trade Type
print("\n📌 Trade Count by Trade Type:")
trade_type_count = system_a_clean.groupby("trade_type")["trade_id"].count()
print(trade_type_count)

# 5. Total NAV by Trade Type
print("\n📌 Total NAV by Trade Type:")
trade_type_nav = system_a_clean.groupby("trade_type")["nav"].sum().round(2)
print(trade_type_nav)

# 6. Active vs Inactive count
print("\n📌 Active vs Inactive Count:")
status_count = system_a_clean.groupby("status")["trade_id"].count()
print(status_count)

# 7. Reconciliation summary
print("\n📌 Reconciliation Summary:")
recon_summary = pd.DataFrame({
    "Metric": [
        "Total Trades",
        "Matched Trades",
        "Mismatched Trades",
        "Match Rate %"
    ],
    "Value": [
        len(comparison),
        len(matches),
        len(mismatches),
        f"{round(len(matches)/len(comparison)*100, 1)}%"
    ]
})
print(recon_summary)

# 8. Mismatch breakdown by Asset Class
print("\n📌 Mismatches by Asset Class:")
mismatch_by_asset = mismatches.groupby(
    "asset_class_sysA")["trade_id"].count()
print(mismatch_by_asset)

# 9. Highest NAV funds
print("\n📌 Top 3 Highest NAV Funds:")
top_nav = system_a_clean.nlargest(3, "nav")[["fund", "asset_class", "nav"]]
print(top_nav)

# 10. Lowest NAV funds
print("\n📌 Bottom 3 Lowest NAV Funds:")
bottom_nav = system_a_clean.nsmallest(3, "nav")[["fund", "asset_class", "nav"]]
print(bottom_nav)

print("\n" + "=" * 40)
print("FINAL SUMMARY")
print("=" * 40)
print(f"Total Trades Processed:     {len(system_a_clean)}")
print(f"Total NAV System A:         {round(system_a_clean['nav'].sum(), 2)}")
print(f"Total NAV System B:         {round(system_b_clean['nav'].sum(), 2)}")
print(f"Total NAV Difference:       {round(system_a_clean['nav'].sum() - system_b_clean['nav'].sum(), 2)}")
print(f"Matched Trades:             {len(matches)} ✅")
print(f"Mismatched Trades:          {len(mismatches)} ❌")
print(f"Match Rate:                 {round(len(matches)/len(comparison)*100, 1)}%")
print(f"Inactive Funds:             {len(system_a_clean[system_a_clean['status'] == 'Inactive'])}")

print("\n✅ Summary Report Complete!")
```
