**Comparing two data frames**

```python
df1 = pd.DataFrame({
    "Fund": ["FundA", "FundB", "FundC"],
    "NAV": [100.5, 200.3, 150.0]
})

df2 = pd.DataFrame({
    "Fund": ["FundA", "FundB", "FundC"],
    "NAV": [100.5, 199.0, 150.0]  # FundB is different!
})
```

```python
# Find mismatches
comparison = df1.merge(df2, on="Fund", suffixes=("_source", "_target"))
comparison["Match"] = comparison["NAV_source"] == comparison["NAV_target"]
print(comparison)
```

```python
#Exporting results
comparison.to_csv("reconciliation_output.csv", index=False)
```
